```python
xml_path_dict = {i.stem: i for i in new_xmls}
png_path_dict = {i.stem: i for i in new_pngs}
```

```python
def get_element_by_tag(xml_path: Path, parent: ET.Element, tag: str, bound_number: int, bound:int=None) -> list[ET.Element]:
    try:
        elements = parent.findall(tag)
        if bound_number:
            if bound == ">":
                assert len(elements) > bound_number
            elif bound == "==":
                assert len(elements) == bound_number
            elif bound == "lt":
                assert len(elements) < bound_number
            elif bound == ">=":
                assert len(elements) >= bound_number
            elif bound == "<=":
                assert len(elements) <= bound_number
            elif bound == "!=":
                assert len(elements) != bound_number 
            elif bound != None:
                print(f"{bound} is not valid. Choose from one of the following ['>', '==', '>=', '<=', '!=']") 
    except:
        print(f"{xml_path} has more or less than one <{tag}> tags from {parent}. Skipping....")
        return
    return elements
```

```python
def get_coords(xml_filename: Path, bounding_boxes: ET.Element) -> tuple[tuple[int, int], tuple[int, int]]:
    xmin = get_element_by_tag(xml_filename, bounding_boxes, "xmin", 1, "==")
    ymin = get_element_by_tag(xml_filename, bounding_boxes, "ymin", 1, "==")
    xmax = get_element_by_tag(xml_filename, bounding_boxes, "xmax", 1, "==")
    ymax = get_element_by_tag(xml_filename, bounding_boxes, "ymax", 1, "==")
    if xmin == None or ymin == None or xmax == None or ymax == None:
        return None
    xmin = int(xmin[0].text)    
    ymin = int(ymin[0].text)
    xmax = int(xmax[0].text)
    ymax = int(ymax[0].text)
    return ((xmin, ymin), (xmax, ymax))
```

```python
BGR_color_dict = {"Red Light": (0, 0, 255), "Blue Light": (255, 0, 0), "Blue Plate": (255, 238, 131), "Red Plate": (42, 76, 234)}
```

```python
i = 0
for xml in new_xmls:
    if xml.stem in png_path_dict:
        png = png_path_dict[xml.stem]
    else:
        print(f"{xml.stem}.xml's corresponding png is not in {IMG_OUT_PATH}")
        continue
    i += 1
    if i >= 5:
        break
    error = False
    img = cv2.imread(str(png))
    try:
        tree = ET.parse(xml)
        root = tree.getroot()
    except:
        print(f"Error in parsing {xml}. Skipping...")
    xml_filename = get_element_by_tag(xml, root, "filename", 1, "==")
    if xml_filename == None:
        continue
    xml_filename = xml_filename[0]
    objects = get_element_by_tag(xml, root, "object", 0, ">")
    if objects == None:
        continue
    for object in objects:
        name = get_element_by_tag(xml, object, "name", 1, "==")
        bounding_boxes = get_element_by_tag(xml, object, "bndbox", 1, "==")
        if name == None or bounding_boxes == None:
            error = True
            break
        name = name[0].text
        if name not in BGR_color_dict:
            print(f"{name} not in BGR_color_dict for {xml}: Skipping image...")
            error = True
            break
        bounding_boxes = bounding_boxes[0]
        coords = get_coords(xml, bounding_boxes)
        if coords == None:
            error = True
            break
        starting_coords = coords[0]
        ending_coords = coords[1]
        cv2.rectangle(img, starting_coords, ending_coords, BGR_color_dict[name], 2)
    fig, ax = plt.subplots()
    plt.imshow(img[:, :, ::-1])
    if error:
        # break
        continue
    # break
```

```python
def change_Bright(img, val):
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    hsv = np.array(hsv, dtype=np.float64)
    val += 100
    val = val / 100  # dividing by 100 to get in range 0-1.5
    # scale pixel values up or down for channel 2(Value)
    hsv[:, :, 2] = hsv[:, :, 2] * val
    hsv[:, :, 2][hsv[:, :, 2] > 255] = 255  # setting values > 255 to 255.
    hsv = np.array(hsv, dtype=np.uint8)
    res = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)
    return res
```
